```python
#Sum values in an array
import numpy as np

L=np.random.random(10)
#Crea array de 10 numeros aleatorios
sum(L)
#lA funcion sum, suma todo los numeros de un arreglo
```

```python
#Maximo y minimo
big_array=np.random.random(1000)
min(big_array), max(big_array)
```

```python
#Agregados multidimensionales
M=np.random.random((5,4))
sum(M) #Suma cada columna
M.sum() #Suma todos los valores
```

```python
#Minimo de cada COLUMNA
M.min(axis=0)
```

```python
#Maximo de cada FILA
M.max(axis=1)
```

```python
#Es mas facil usar .inset_axes para crear subgrauficas, me quito muchis errores
import numpy as np
import matplotlib.pyplot as plt

# Crear la figura y los ejes
fig = plt.figure(figsize=(10, 6))
plt.style.use('classic')

# Gráfico principal de senos
ax1 = plt.subplot(1, 2, 1)
for i in range(0, 10):
    x = np.linspace(0, 25, 1000)
    ax1.plot(x, np.sin(x + i))
ax1.set_title('Gráfico de Senos')

# Subgráfico de coseno en el gráfico principal de senos
x_cos_sub = np.linspace(0, 10, 20)
ax2 = ax1.inset_axes([0.5, 0.35, 0.2, 0.2])
ax2.plot(x_cos_sub, np.cos(x_cos_sub))
ax2.set_title('Subgráfico de Coseno')

# Gráfica de coseno a la derecha
ax3 = plt.subplot(1, 2, 2)
x_cos_extra = np.linspace(0, 10, 1000)
ax3.plot(x_cos_extra, np.cos(x_cos_extra), color='orange')
ax3.set_title('Gráfica Extra de Coseno')

plt.tight_layout()
plt.show()
```
