```python

# Instala la biblioteca pandas si no está instalada
!pip install pandas

import pandas as pd

# Cargar el archivo Excel desde tu computadora
excel_file_path = '/content/DataW.xlsx'  # Reemplaza esto con la ruta real del archivo Excel
data = pd.read_excel(excel_file_path)

# Extrae los datos de las columnas "Year" y "ERA5" en forma de arrays
year_array = data['Year'].values
era5_array = data['ERA5'].values

```

```python


# Datos de las columnas "Year" y "ERA5"
x = year_array
y = era5_array

n = len(x)  # Calculamos la longitud de x para saber entre qué valores promediar
xy = x * y

# Calculamos la pendiente (m) y la ordenada al origen (b)
m = (n * sum(xy) - sum(x) * sum(y)) / (n * sum(x**2) - sum(x)**2)
b = (sum(y) - m * sum(x)) / n

# Calculamos la línea de regresión
regression_line = m * x + b

# Crear el gráfico
plt.scatter(x, y, label='Datos')
plt.plot(x, regression_line, color='red', label='Línea de regresión')
plt.xlabel('Year')
plt.ylabel('ERA5')
plt.title('Regresión Lineal')
plt.legend()
plt.show()

print("Pendiente (m):", m)
print("Ordenada al origen (b):", b)

```
