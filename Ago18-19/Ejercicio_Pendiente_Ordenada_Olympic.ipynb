```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/olympic100m_men.txt",delimiter = ",").T
x=data[0]
y=data[1]
plt.scatter(x,y)


```


El método de mínimos cuadrados es una técnica utilizada para encontrar la mejor aproximación lineal de un conjunto de datos. Supongamos que tenemos \(n\) puntos \((x_i, y_i)\) y queremos encontrar la línea recta \(y = mx + b\) que mejor se ajuste a estos puntos minimizando la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos.

La ecuación de la línea general es: \( y = mx + b \)

Nuestro objetivo es encontrar los valores de \(m\) y \(b\) que minimizan la función de error cuadrático:
$$ E(m, b) = \sum_{i=1}^{n} (y_i - (mx_i + b))^2 $$

Para encontrar los valores de \(m\) y \(b\) que minimizan \(E(m, b)\), tomamos las derivadas parciales de \(E\) con respecto a \(m\) y \(b\) y las igualamos a cero:
$$ \frac{\partial E}{\partial m} = -2 \sum_{i=1}^{n} x_i(y_i - (mx_i + b)) = 0 $$
$$ \frac{\partial E}{\partial b} = -2 \sum_{i=1}^{n} (y_i - (mx_i + b)) = 0 $$

Resolviendo estas ecuaciones, obtenemos las fórmulas para \(m\) y \(b\):
$$ m = \frac{n(\sum_{i=1}^{n} x_i y_i) - (\sum_{i=1}^{n} x_i)(\sum_{i=1}^{n} y_i)}{n(\sum_{i=1}^{n} x_i^2) - (\sum_{i=1}^{n} x_i)^2} $$




```python
n=len(x)      #Calculamos la longitud de x para saber entre que valor promediar
m = (n * sum(xy) - sum(x) * sum(y)) / (n * sum(x**2) - sum(x)**2)
m
```

Ahora calculamos b
$$ b = \frac{\sum_{i=1}^{n} y_i - m\sum_{i=1}^{n} x_i}{n} $$

```python
b = (sum(y)-m*sum(x))/n
b
```

Realizamos la grafica correspondiente

```python
Regresion_lineal = m*x+b
plt.scatter(x,y,Regresion_lineal)
plt.plot(x, Regresion_lineal, color='red', label='Línea de regresión')
```
